```python
import json, os
```

```python
json_folder = '/home/maaz-lfd/Maaz/Thesis/Thesis/dataset/clinical_trials/pmc'
txt_folder = '/home/maaz-lfd/Maaz/Thesis/Thesis/dataset/clinical_trials/pmc_text'
```

```python
for file in os.listdir(json_folder):
    
    print(file)
    data = json.load(open(os.path.join(json_folder,file)))
    
    data_txt = f"PAPER ID: {data['paper_id']}\n\nTITLE:\n{data['metadata']['title']}\n\nABSTRACT:\n{data['metadata']['abstract']}"
    prev_section = ''
    for i in data['body_text']:
        
        if i['section']==prev_section:
            data_txt += '\n'+i['text']
        else:
            data_txt += f"\n\n{i['section']}:\n{i['text']}"
            prev_section = i['section']
    file_rename = file.replace('.xml.json','.txt')
    with open(os.path.join(txt_folder,file_rename),"w") as f:
        f.write(data_txt)
```
