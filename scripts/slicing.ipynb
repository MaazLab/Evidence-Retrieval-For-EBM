```python
import os
import pandas as pd
import shutil
```

```python
gs_df = pd.read_csv('/home/maaz-lfd/Maaz/Thesis/Thesis/dataset/clinical_trials/GS_paper_metadata.csv')
gs_df
```

```python
pmcid = gs_df['pmcid'].to_list()
```

```python
pmcid
```

```python
folder_path = "/home/maaz-lfd/Maaz/Thesis/Thesis/dataset/labelled dataset"
```

```python
for i in os.listdir(os.path.join(folder_path,"original")):
    if not i.endswith('.xlsx'):
        for file in os.listdir(os.path.join(folder_path,"original",i)):
            if file[:-4] in pmcid:
                dir_path = os.path.join(folder_path,'sliced',i)
                os.mkdir(dir_path) if not os.path.exists(os.path.join(folder_path,'sliced',i)) else None
                shutil.copy(src=os.path.join(folder_path,'original',i,file),dst=os.path.join(folder_path,'sliced',i,file))
                


```

```python
count = 0
for dir in os.listdir(os.path.join(folder_path,'sliced')):
    count += len(os.listdir(os.path.join(folder_path,'sliced',dir)))
```
